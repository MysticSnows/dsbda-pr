```python
from scipy import stats
```

```python
# measure of how far a particular data point deviates 
# from the mean of the distribution
z = np.abs(stats.zscore(df['math score']))
```

```python
print(z)
```

```python
threshold = 0.18
sample_outliers = np.where(z < threshold)
sample_outliers
```

```python
sorted_rscore = sorted(df['reading score'])
sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
r_outliers = []  # Initialize an empty list to store outliers

for i in sorted_rscore:
    if (i < lwr_bound or i > upr_bound):
        r_outliers.append(i)  # Add outliers to the list

print(r_outliers)  # Print the list of outliers
```
